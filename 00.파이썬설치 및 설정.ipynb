#### 4. 강제중단 / 재 실행

<img src="./images/00.파이썬설치 및 설정_03_강제중단.PNG" width="600" height="300" />

* 제목줄 아래에 Kernel메뉴가 있다. 이 커널은 iPython(Jupyter Notebook)대화창 아래에서 백그라운드와 비슷하게 실행되는 일종의 운영체제와 같은 개념이다.

* Kernel의 모든 메뉴는 코드를 삭제하지 않는다.
 * Interrupt : 실행중인 코드를 강제로 중지. 에러메시지가 뜨면서 실행이 중지가 된다.
 * Restart : 실행중인 코드가 중지가 되며 재시작된다. 코드나 실행결과는 삭제가 되지 않는다.
 * Restart & Clear Output : 코드가 중지가 되며, 실행결과는 삭제가 된다.
 * Restart & Run All : 재 시작후에 모든 셀의 코드를 위에서 부터 순차적으로 실행된다.
 * Reconnect : 인터넷 연결이 끊어 졌을 때 연결을 재시도 한다.
 * Shutdown : 커널을 종료한다. 이 메뉴는 실행결과는 삭제되지 않지만 완전 종료가 되는 상태로 메모리에서 해제가 된다.
 
* Shutdown이 되었거나 인터넷 연결이 끊어 졌거나 기타 문제가 있다면 탭에 알림표시가 나타난다.

### E. Jupyter Notebook의 기능

#### 1. DocString표시

선언한 변수 뒤에 <code>?</code>를 붙여서 셀을 실행하는 것으로 해당 변수의 상태를 확인할 수가 있다.
또한 변수를 선언한 후에 <code>Shift + Tab</code>를 누르면 Tool Tip이 표시가 된다. Tool Tip에는 DocString의 일부 내용이 표시가 된다.

변수 또는 함수들의 자세한 내용을 보기 위해서는 <code>help()</code>함수를 사용한다.

#### 2. Image첨부하기

그림파일 첨부하기 위해서는 <code>img tag</code>를 이용하거나 Drag & Drop하면 된다.

#### 3. shell(명령 프롬프트)의 이용

cmd창에서 쓰는 명령을 그대로 사용하는데 맨 앞에 <code>!</code>를 붙여서 사용한다
예를 들어 <code>!dir</code> 또는 <code>!cd documents</code>형태로 Dos명령을 사용한다.


#### 4. 매직명령어
