```python
# %(매직명령어)
# %pwd : 현재 디렉토리 보기
# %ls : 현재 디렉토리 목록
# %who: 현재 세션에서 사용중인 변수나 함수, 패키지등의 목록을 리스트로 리턴
# %hist: 명령어의 히스토리를 리스트로 반환
# %matplotlib : 그래프출력을 즉시 웹에 출력여부를 결정
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib

%who
```

```python
# 1. 난수발생
data = np.random.rand()
print(data)
print(type(data))

data = np.random.rand(50) # 50개의 난수를 한번에 발생
print(data)
print(type(data))

# ?np.random
# ?np.random.rand
```

```python
# 2. pandas : 1차원자료를 생성
# ?pd.Series
s = pd.Series(data)
print(s)
print(type(s))
print(dir(s))
```

```python
# pandas 그래프그리기 : 생성된 자료(ndarray)를 가지고 시각화
s.plot()
#?s.plot
```

```python
# 3. numpy : 2차원자료를 생성
# numpy로 랜덤하게 10행 3열의 2차원배열(정확하게는 행렬이라고 한다.)
data_set = np.random.rand(10,3)
print(data_set)
```

```python
# 4. pandas : 2차원 자료구조 (DataFrame)
#pd.DataFrame?
df = pd.DataFrame(data_set)
df
print(df)
print(type(df))

df1 = pd.DataFrame(data=data_set, columns=['2017', '2018', '2019'])
df1
```

```python
# 시각화 : pandas로 2차원 그래프 그리기(line그래프)
# pandas.plot()는 기본이 라인그래프
#df1.plot?
df1.plot()
```

```python
# bar 그래프
df1.plot(kind='bar')
```

```python
# barh 그래프
df1.plot(kind='barh')
```

```python
# area 그래프
df1.plot(kind='area')
```

```python
# area 그래프 - stacked
df1.plot(kind='area', stacked=False)
```

```python
# pie chart
# 5개의 난수를 발생, Series자료형의 인덱스값
ser = pd.Series(data=np.random.rand(5),
                index=['2015','2016','2017','2018','2019'], 
                name='Sales Amount')
print(ser)
```

```python
# 시각화 : pichart
# autopct : 소숫점 2자리로 설정
# figsize(x,y) : x축과 y축의 그래프 크기를 설정
# fontsize: 폰트크기
ser.plot(kind='pie', autopct='%.2f%%', figsize=(10,5), fontsize=20)
```
