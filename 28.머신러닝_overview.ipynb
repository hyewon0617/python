### 기술통계와 모델링

##### 1. Data Mining

* 대용량의 데이터로부터 유용한 정보를 캐내는(mining) 작업
* 대용량 데이터에서 존재하는 데이터간의 관계, 패턴, 규칙등을 찾아내고 모형화해서 기업의 경쟁력 확보를 위한 의사결정을 돕는 일련의과정
* 데이터 마이닝은 큰 데이터 집합에서 동작 가능한 정보를 찾는 프로세스로서 데이터마이닝에서는 수학적인 분석을 사용하여 데이터에 있는 패턴이나 추세등의 정보를 찾는다. 일반적으로는 관계가 매우 복잡하거나 데이터가 너무 많아서 전통적인 데이터 탐색 또는 분석하기위한 패턴을 찾을 수가 없게 되었다.
* 이러한 패턴과 추세를 수집하여 데이터마이닝 모델로 정의할수가 있는데 마이닝 모델을 다름과 같은 특정 시나리오에 적용할 수 있다.
 - 예측 : 서버부하 또는 서버가동중지 시간을 예측하는 추정
 - 위험 및 확률 : 타켓 메일링을 위한 최적합 고객의 선택, 위험 시나리오를 위한 예산손익에 대한 결정, 진단 및 기타결과에 확률을 할당
 - 추천 : 함께 판매될 가능성이 높은 제품을 결정, 추천제품 구성
 - 연계성찾기 : 장바구니의 고객선택 분석, 적절한 다음 이벤트 예측
 - 그룹화 : 고객 또는 이벤트를 관련된 항목의 클러스터로 분리, 선호도 분석 및 예측

1. CRISP-DM(CRoss Industry Standard Processing for Data Mining)

 - 데이터마닝을 위한 업계표준 프로세스 : 데이터마이닝작업을 위한 검증된 방법론
   
   1) 비지니스이해(Business Understanding)
   
      각종 참고자료와 현업 책임자와 의사소통을 통해 해당 비즈니스를 이해 하는 단계
      반드시 그 분야의 전문가와 함께 참여해서 진행된다.
      
   2) 데이터의 이해(Data Understanding)
      
      초기 데이터를 수집, 데이터에 대한 이해, 데이터 품질검증, 자료건수, 자료의 속성,
      자료의 품질등 현업에서 보유 및 관리하고 있는 데이터를 이해하는 단계   
   
   3) 데이터준비(Data Preperation)
   
      모델링에 필요한 데이터 변환, 정제등의 작업 -> 데이터의 정제, 새로운 데이터생성,
      데이터 업데이트등 자료를 분석가능한 상태로 만드는 단계
      
   4) 모델링(Modeling)
   
      데이터 분석을 위한 모델링 기법을 선택하고 모델을 생성하는 단계
      
   5) 평가(Evaluation)
   
      분석 결과에 대한 분석 및 평가, 모델의 해석가능여부, 새로운 자료에도 잘 적용(예측)이
      되는지 여부를 검토하는 단계. 필요할 경우 모델을 재구축
      
   6) 배포(Deployment)
 
      최종 보고서 작성 및 배포단계, 각 관리자에게 전달하여 필요한 조치를 취하는 등 검토가
      완료된 모델을 실제 현업에 적용하는 단계
 
 - 참고 : https://www.ibm.com/support/knowledgecenter/ko/SS3RA7_18.0.0/modeler_crispdm_ddita/clementine/crisp_help/crisp_overview.html
 
 
2. 머신러닝(인공지능) / 데이터마이닝의 다양한 기법

   1) 분류모형(Classification Models)
    
      특정 기준(정답, 지도학습)에 의해 분석대상을 2개 또는 3개이상의 집단으로
      분류하는 예측모형(부도예측, 기상예측, 채권등급예측 등)
      
      a. 지도학습 : 미리 답을 지정하는 학습(신규고객의 대출시 연체여부에 따라 대출제한등..
      b. 비지도학습 : 답이 없이 데이터를 분석해서 결과를 도출하는 것
      c. 알고리즘
         ... 다중판별분석(MDA, Mutiple Discriminant Analysis)
         ... 로직스틱 회귀분석(LOGIT, Logistic Regression)
         ... 인공신경망(ANN, Artificial Nural Networks) -> 더 발전된 것이 Deep Learnning
         ... 사례기반추론(CBR, Case-Based Reasoning)
         ... 의사결정나무(DT, Decision Treens)
         ... SVM(Support Vetor Machines)
   
   2) 최적화기법(Optimization Methods)
   
      주어진 제약조건하에서 특정 목적함수를 최대, 최소화하는 변수들의 최적의 값을 도출
      하는 기법이다. 예를 들어 공장의 생산량 최대화의 문제, 비용을 최소화하는 유통경로등
      
      a. 알고리즘
         ... 선형계획법(LP, Linear Programming)
         ... 유전자 알고리즘(GA, Genetic Algorithms)
      
   3) 가치평가기법(Valuation Methods)
   
      정성적 측정대상에 대한 가치를 비교, 평가하는 기법(차세대전투기선정, 신행정수도선정 등)
      
      a. 알고리즘
         ... 분석적 계층 프로세스(AHP, Analytic Hieracht Process)
         ... 분석적 네트워크 프로세스(ANP, Analytic Network Process)
         ... 로포락분석(DEA, Data Envelopment Analysis)
      
   4) 군집화기법(Clustering Methods)
   
      사전에 정해진 기준없이 동질의 데이터드들을 같은 그룹으로 묶어주는 기법(고객세분화등)
      
      a. 알고리즘
         ... K-means 분류기법(K-means clustering)
      
   5) 전처리기법(Preprocessing Methods)
   
      예측모형의 성과를 향상시키기 위해 입력데이터에 대한 사전 처리를 수행하는 기법
      
      a. 알고리즘
         ... 주성분분석(PCA, Principal Component Analysis)
         ... 퍼지이론(Fuzzy Theory)

##### 2. 데이터전처리

1. 결측값(Missing Value)의 처리

    너무 많은 항목이 비어있거나 너무 많은 항목이 비어있는 레코드등의 자료를 삭제하거나
    기타 나머지 항목에대해서는 일정의 값으로 대체하는 처리를 하는 것
    
    평균값(Mean)/중앙값(Median)/최빈치(Mode)
    예) 1,2,3,4,5,6,6,6,7,8
    평균값 : (1+2+3+4+5+6+6+6+7+8) / 11 = 4.727
    중앙값 : 5
    최빈치 : 6

2. 정성적인 자료의 정량화

   각 속성은 단일변수(automic value)를 갖도록 수정하는 것을 말한다. 정성적인 변수의 경우
   0과 1의 binary code로 변환해야 추후 해석이 가능하다.
   예를 들어 `주소의 변환`, `성별의 변환`등이 있다.

3. 이상치(Outlier)의 제거

   상식적으로 말이 않되거나 잘못 입력된 것으로 추정되는 변수값을 조정해서 일괄적으로
   상위 10%와 하위 10%에 해당하는 값들을 단일값으로 부여하는 경우도 있다.
   예를 들어 체중80kg이상은 무조건 80kg로 체중 40kg이하는 무조건 40kg으로 부여

4. 새로운 파생변수의 개발
   
   기존의 변수를 조함하여 새로운 변수를 개발하는 것을 말한다. 본래는 비율변수인 값을 
   의미가 있는 정보로 구간화하여 새로운 명목변수로 만드는 것등
   
5. 정규화(Normalization)

   모든 입력변수의 값이 최소 0에서 1사의 값을 갖도록 조정하거나 평균 0을 갖는 표준 정규
   분포를 갖도록 조정하는 것
   
   정규화공식(Min-Max Normalization) :  (x - 최소값) / (최대값 - 최소값)
   
   예를 들어 전체 고객중 체중이 가장 작은 사람이 40kg, 가장 큰 사람이 120kg라고 하면
   40Kg을 0, 120Kg을 1로 변환 후 80Kg은 (80-40) /(12-40) = 0.5로 변환
   
6. 자료의 구분

   1) 과적합화(Overffiting)가 발생 가능성
   
      다음날의 주가지수를 예측하는 모형이 A, B가 있다면
      A모델은 구축한 날까지의 주가(과거주가)는 99.99%를 맞추는 반면에 
      그 다음날의 주가지수를 예측하면 70%정도 맞추었다.
      B모델을 과거주가는 83%정도 예측하는데 미래주가는 80%정도 예측한다면
      A, B모델 중 더 잘 구축된 모델은? -> 주가는 미래예측이므로 B모델이 더나은 모델이다.
      
   2) 과적합화의 예방법 : 모델 구축시에 Hold-out data의 개념을 도입
      
      Hold-out data(검증) : 모델이 일반성을 갖는지 확인하기 위해 남겨누는 unknown data
      통계모형을 구축할 때 전체데이터가 100이라면 학습용데이터와 검증용데이터를 나누어서
      학습용:검증용=8:2 또는 7:3의 비중으로 자료를 미리 나누어 두는 방법   
      
   3) 0/1의 예측의 경우 0과 1의 비율이 각 데이터셋마다 1:1의 비중이 되도록 섞어야 한다.

7. 모형에 들어갈 후보 입력값의 선정

   1) 카이제곱검증(Chi-square test) : 성별(이산형), 구매여부(이산형)
   2) 독립표본 t검정(t-test) - 이분류모형경우에 사용, 구매여부(이산형), 체중(연속형)
   3) 분산분석(ANOVA) - 다분류모형의 경우에 사용, 고객등급(이산형), 체중(연속형)

##### 3. 4가지 종류의 척도

   명목척도, 서열척도 -> 이산형변수
   등간척도, 비율척도 -> 연속형변수
   
   1) 명목척도(Nominal Scales) : 단순히 번호만 부여한 척도
   
      (1) 정의 : 대상의 특성을 분류하거나 확인할 목적으로 사용하는 척도
      (2) 예시 : 축구선수 등번호, 입시에 사용되는 출신학교 코드
   
   2) 서열척도(Ordinal Scales) : 
   
      (1) 정의 : 대상의 특성이 순서를 밝히기 위해 사용하는 척도(양적비교불가)
      (2) 예시 : 수학시험석차, 선호도 우선순위
      
   3) 등간척도(Interval Scales)
   
      (1) 정의 : 부여된 순위 사이의 간격이 동일한 척도
      (2) 예시 : 온도
      
   4) 비율척도(Ration Scales)
   
      (1) 정의 : 순위 사이의 간격이 동일하고 비율계산이 가능한 척도
      (2) 예시 : 무게, 투표율

##### 4. 통계학의 개요

   1) 통계 및 통계분석
   
      (1) 통계학(Statistics)이란?
          현황을 가능한 정확히 알아내고 미래에 대한 적절한 예측을 하기 위해
          자료를 효율적으로 수집정리하고 분석하는 학문 분야
          
      (2) 통게분석의 목적
          현황을 제대로 파악하기 위해
          미래를 예측하기 위해
          
      (3) 통계학이 적용되는 대표적인 예들
          공학자 : 현재 생산되는 제품의 품질이 어떤 상태에 있는가?
          경영자 : 제품에 대한 소비자의 선호도가 어떻게 나타나고 있는가?
          사회과학자 : 정치에 대한 국민들의 의견은 어떻게 나타나고 있는가?
          
   2) 모집단과 표본
      국회의원들이 소유하고 있는 모든 주택의 평균가격은 얼마인가?
      
      (1) 모집단(population)
          관심의 대상이 되는 집단 전체(예:국회의원소유주택수)
          모수(parameter) : 모집단이 가지고 있는 특징을 나타내는 수치(예:평균가)
          전수조사(census) : 모집단의 모든 개체를 전부 조사하는 방법, 모수의 값
             을 가장 정확하게 알아내는 방법. 단점은 시간과 비용문제가 있다.
      
      (2) 표본(Sample)
          모수의 값을 알아내기 위해 추출된 모집단의 일부분
          일반적인 통계분석에서는 모집단의 일부만 조사하여 모집단의 모수를 추측
          100% 정확하지는 않지만, 거의 근접한 값을 제공해 줄 수 있다.
          대신 시간과 비용을 크게 절약할 수 있는 방법
          
          
      (3) 표본추출
          표본이 모집단을 대표할 수 있도록 추출하는 것이 중요
          표본추출이 잘못되어 발생한 사례
             --> 1936년 미국대선 : 민주당 Franklin Roosevelt vs 공화당 Alf London
                 전화조사 결과 공화당 승리예상. 실제로는 민주당 승리
                 당시에는 전화보급율이 매우 낮았음
          
   3) 통계학의 2가지 분야
   
      (1) 기술통계학 : 수집된 자료들이 가지고 있는 정보를 쉽게 파악할 수 있도록
          표현하는 방법에 대한 내용을 다루는 학문분야로 주어진 자료의 특성을 
          그대로 기술하는 것(예: 매출액, 사원수, 연령별인두등.._
          
          예 : 다차원척도법, 컨조인트분석, 군집분석, 요인분석이 기술통계학에 해당
               (다컷군요!!로 암기)
          
      (2) 추리통계학 : 수집된 자료들을 분석해 볼 때 알아내고자 했던 모집단의 모수
          는 얼마로 추정되는지, 미래예측치는 얼마로 예상되는지 등의 결론을 도출
          하는 학문 분야. 수집된 자료를 통하여 원래 알아보고자 했던 문제에 대한 
          결혼을 도출, 통계학의 대부부 내용들이 여기에 해당한다.
          
          표본의 특성으로 부터 모집단의 특성을 추정하는 것
          
          예) 기술통계학 이외의 분석방법이 추리통계학 분햐
          
   4) 통계분석의 절차
   
      (1) 문제의 올바른 파악 : 대상이 되는 모집단에 대해 정확하게 설정하고
          알아내려는 사항이 무엇인지 제대로 파악
          
      (2) 표본의 추출과 자료의 수집 : 비표본오차(nonsampling error)가 발생하지
          않도록 설계
          표본오차 : 모집단을 표본으포 추정하는데서 방생되는 오차(회피불가)

##### 1. 카이제곱검정

```python
import pandas as pd
import numpy as np
from scipy import stats # 과학기술 통계분석함수 모음
# !pip show scipy
```

```python
data1 = [4,6,17,16,8,9]     # 관측치
data2 = [10,10,10,10,10,10] # 기대치

# data1과 data2사이에 유의한 관계가 있는가?

# 귀무가설 : data1과 data2가 차이가 없다.
# ... 귀무가설(H0) 또는 영가설은 통계학에서 처음부터 버릴 것을 예상하는 가설이다.
# ... 차이가 없거나 의미있는 차이가 없는 경우의 가설이며 이것이 맞거나 맞지않다는
#     통계학적 증거를 통해 증명하려는 가설이다.
#     예를 들어 범죄사건에서 용의자가 있을 때 형사는 이 용의자가 범죄를 저질렀다는
#     추정인 대립가설을 세우게 된다. 이때 귀무가설은 용의자는 무죄라는 가설이다. 
#     통계적인 방법으로 가설검정을 시도할 때 사용된다.
# 대립가설 : data1과 data2가 차이가 있다.
# ... 가설검정이론에서 대립가설(기호 H1) 또는 연구가설, 유지가설이라고도 한다.
# ... 귀무가설에 대립하는 명제

# 검정통계량과 p-value
# stats.chisquare?
chis = stats.chisquare(data1,data2)
print(type(chis))
print(chis)

# 결과분석
# pvalue가 0.05보다 작으므로 95%신뢰하에서 유의한 관계가 있다.
# 귀무가설을 기각, 대립가설을 채택 즉, 두 데이터간의 상관관계는 "차이가 있다"
# statistic = 검정통계량
# pvalue = 피어슨값(상관계수) -> 중요한 값
```

#### 2. 독립(단일)표본 t검정

```python
# 전체학생들중 20명의 학생들을 선택하여 전체학생들의 평균키가 175cm인지 여부를 검정
# 귀무가설 : 학생들의 평균키가 175이다.
# 대립가설 : 학생들의 평균키가 175가 아니다.

# 랜덤시드설정(동일한 난수가 생성) 즉, 랜덤값을 고정
np.random.seed(1)

# np.random.normal(0, 5) : 평균이 0, 표준편차가 5인 난수를 발생
heights = [180 + np.random.normal(0, 5) for i in range(20)]
#rint(heights)

result = stats.ttest_1samp(heights, 175) # 단일표본 t검정
#rint(type(result))
print('검정통계량: %.3f, p-value : %.3f' % result)

# 결과분석
# pvalue가 0.01보다 작으므로 99%신뢰수준하에서
# 학생들의 평균키는 통계적으로 유의하게 차이가 난다고 할 수 있다.
# 따라서 귀무가설을 기각하고 대립가설을 채택한다.
# 즉, 학생들의 평균키는 175가 아니다.
```

#### 3. 독립표본 t검정

```python
# 그룹1과 그룹2에서 각각 20명의 학생들을 선택하여 평균키가 같은 다른지를 검정
# 귀무가설 : 학생들의 평균키가 같다.
# 대립가설 : 학생들의 평균키는 같지 않다 (내가 주장하고 싶은 명제)

np.random.seed(1)

# 그룹1 : 평균 170 표준편차 5
group1 = [170 + np.random.normal(0, 5) for i in range(20)]
# 그룹2 : 평균 175 표준편차 10
group2 = [175 + np.random.normal(0, 10) for i in range(20)]

# 등분산을 가정할 경우(기본값)
result1 = stats.ttest_ind(group1, group2) # 독립표본 t검정
print('검정통계량: %.3f, p-value : %.3f' % result1)

# 등분산을 가정하지 않을 경우
result2 = stats.ttest_ind(group1, group2, equal_var=False)
print('검정통계량: %.3f, p-value : %.3f' % result2)

# 결과분석
# p-value < 0.05 : 95%신뢰수준하에서 유의
# p-value < 0.01 : 99%신뢰수준하에서 유의
# p-value가 0.05보다 작으므로 95% 신뢰수준하에서는 유의하지만 99% 신뢰하에서는 무의하다.
# 두 그룹 학생들의 평균키는 통계적으로 유의하게 차이가 난다고 할 수 있다.
# 따라서 귀무가설을 기각하고 대립가설을 채택한다.
# 즉, group1과 group2의 학생들의 평균키는 같지가 않다.
print("그룹1 학생들의 평균키 = ", np.mean(group1))
print("그룹2 학생들의 평균키 = ", np.mean(group2))
```

#### 4. 대응표본 t검정

>독립표본 t검정은 두개의 검정대상이 별개의 그룹이었다면 대응표본 t검정은 동일그룹이다.

```python
# 다이어트약을 복용한 사람들중에서 20을 채택하여 복용전후의 체중차이가 유의미한지 검정
# 귀무가설 : 다이어트약을 복용한전후에 체중의 차이가 없다.
# 대립가설 : 다이어트약을 복용한전후에 체중의 차이가 있다.

np.random.seed(1)

# 복용전 평균 60kg, 표준편차 5
before = [60 + np.random.normal(0, 5) for _ in range(20)]

# 복용후 평균 = 복용전체중 x 0.99, 표준편차 0.02
after = [w + np.random.normal(0.99, 0.02) for w in before]

#print(before)
#print(after)

# 대응표본 t검정
result = stats.ttest_rel(before, after)
print('검정통계량: %.3f, p-value : %.3f' % result)

# 결과분석
# p-value가 0.05보다 작으므로 95% 신뢰수준하에서는 유의하다
# 다이어트약 복용전후의 체중은 통계적으로 유의하게 차이가 난다고 할 수 있다.
# 따라서 귀무가설을 기각하고 대립가설을 채택한다.
# 즉, 다이어트약 복용전후의 체중차이가 있다.
```

#### 5. 아노바분석(분산분석 Analysis of Variance)

어떤 회사에 김부장은 4개의 각기 다른 신입사원 교육훈련 기법의 효과성을 평가하고자 한다. 
새로 입사한 32명의 신입사원에거 4가지 기법을 임의로 적용시켜 교육을 실시한 후에 한달간의 훈련기간이 끝난후에 표준시험결과 4개의 교육훈련기법간 차이가 있는지? 만약 있다면 어떻게 다른지를 검정할 때 3개 이상의 대응표본을 비교해야 하므로 `일원배치분산분석기법`을 사용해야 한다.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. 한달간 교육결과 - 무의한 샘플데이터
a = [66,74,82,75,73,97,87,78]
b = [76,73,81,79,72,91,83,79]
c = [69,72,81,78,72,91,89,71]
d = [61,72,83,74,75,96,87,79]

# 2. 한달간 교육결과 - 유의한 샘플데이터
# a = [66,74,82,75,73,97,87,78]
# b = [72,51,39,62,74,64,78,63]
# c = [61,60,57,60,81,55,70,71]
# d = [63,61,76,84,58,65,69,80]
```

```python
# 귀무가설 : 4개의 교육훈련 기법간의 차이가 없다.
# 대립가설 : 4개의 교육훈련 기법간의 차이가 있다.

plot_data = [a,b,c,d]
ax = plt.boxplot(plot_data)
plt.show()

# 일원배치 분산분석(아노바분석)
# stats.f_oneway?
F_statistic, pValue = stats.f_oneway(a,b,c,d)
print('검정통계량: %.3f, p-value : %.3f' % (F_statistic, pValue))
```

```python
# 2. 한달간 교육결과 - 유의한 샘플데이터
a = [66,74,82,75,73,97,87,78]
b = [72,51,39,62,74,64,78,63]
c = [61,60,57,60,81,55,70,71]
d = [63,61,76,84,58,65,69,80]
```

```python
plot_data = [a,b,c,d]
ax = plt.boxplot(plot_data)
plt.show()

# 일원배치 분산분석(아노바분석)
F_statistic, pValue = stats.f_oneway(a,b,c,d)
print('검정통계량: %.3f, p-value : %.3f' % (F_statistic, pValue))
```

###### get_rdataset 함수
statsmodels의 datasets서브패키지에 있는 `get_rdataset()`함수를 사용하면 표준데이터셋을 가져올 수 있다. 이 함수의 사용법은 `get_rdataset(item, [package="datasets"])`와 같이 사용하고 패키지 이름이 `datasets`일 경우에는 생략할 수 있다.
이 함수를 사용하려면 인터넷에서 데이터를 다운로드받기 때문에 인터넷에 연결되어 있어야 한다.
이렇게 받은 데이터는 아래와 같은 속성을 가지고 있다.
>* package : 데이터를 제공하는 R패키지 이름
* title : 데이터의 이름
* data : 데이터를 담고 있는 데이터프레임

`__doc__`은 데이터에 대한 설명 문자열로서 이 설명은 R패키지의 내용을 그대로 가져온 것이므로 예제코드가 R로 되어 있어서 파이썬에서는 사용할 수 없다.

```python
# 4. 호흡기 질환 사망자 수(영국)
# MASS패키지의 death데이터는 1974~1979년 사이의 호흡기 질환 사망자 수를 나타내는
# 시계열 데이터이다.
data = sm.datasets.get_rdataset("deaths", "MASS")
df = data.data
display(df.head())
print(data.__doc__)
```

```python
# 이 시계열 데이터에서는 시간이 1년을 1.0으로 1개월을 1/12로 하는 값(year-fraction)으로
# 인코딩되어 있다. 이 값을 파이썬의 datetime포맷으로 바꾸기 위해 아래와 같은 함수를 
# 적용해야 한다.
def yearfraction2datetime(yearfraction, startyear=0):
    import datetime
    import dateutil
    year = int(yearfraction) + startyear
    month = int(round(12 * (yearfraction - year)))
    delta = dateutil.relativedelta.relativedelta(months=month)
    date = datetime.datetime(year, 1, 1) + delta
    return date

display(df.tail())
df["datetime"] = df.time.map(yearfraction2datetime)
df.tail()
```

```python
df.plot(x="datetime", y="value")
plt.title(data.title)
plt.show()

# 결과분석
# 이 데이터는 계절에 따른 패턴 즉, 계절성(seasonality)를 보이는 시계열데이터이다.
```

```python
# 5. 항공운송인원
# datasets패키지의 AirPassengers데이터는 1949~1960년 사이의 국제 항공운송인원을 나타내는
# 시계열데이터이다.
data = sm.datasets.get_rdataset("AirPassengers")
df = data.data
display(df.head())
print(data.__doc__)
```

```python
display(df.tail())
df["datetime"] = df.time.map(yearfraction2datetime)
df.tail()
```

```python
df.plot(x="datetime", y="value")
plt.title(data.title)
plt.show()

# 결과분석
# 계절성과 증가추세(trend)를 동시에 보이고 있다.
```
